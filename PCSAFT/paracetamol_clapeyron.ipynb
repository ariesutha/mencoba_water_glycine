```julia
model = CompositeModel(["water","paracetamol"];fluid=PCSAFT,solid=SolidHfus)
```

```julia
T = LinRange(300.,350.,100)
p = 1e5 

s = zeros(length(T))

for i in 1:length(T)
    s[i] = sle_solubility(model,p,T[i],[1.,1.];solute=["paracetamol"])[2]
end

plt.clf()
plt.plot(T,s)
plt.xlabel("Temperature (K)")
plt.ylabel("Solubility (mol/mol)")
plt.ylim(0,1e-4)
plt.xlim(300,350)
display(plt.gcf())
```
